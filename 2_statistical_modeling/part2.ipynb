# Modelamiento con mtodos no-paramtricos

En esta leccin revisaremos dos mtodos no-paramtricos de modelamiento, los histogramas y la estimacin de densidad por *kernels* (Kernel Density Estimation, KDE). 

Estos mtodos no-paramtricos nos permiten modelar distribuciones arbitrarias, es decir distribuciones cuya forma funcional no es conocida a priori.

Por lo anterior, estos mtodos son ideales para procesos de exploracin, donde buscamos estudiar las caractersticas estadsticas de nuestros datos. Nuestras observaciones luego puede ser insumo para proponer modelos paramtricos (estructura conocida a priori) ms robustos.

## El Histograma

Un histograma es una representacin aproximada de una distribucin de datos numrica. Esta aproximacin se realiza utilizando el concepto de frecuencia (conteo). En su forma ms simple la construccin de un histograma tiene tres pasos

- Establecer un rango para el histograma: valor mnimo y mximo
- Dividir el rango en un cierto nmero de particiones (*bins*)
- Medir la frecuencia relativa en cada bin, es decir la cantidad de datos que caen en dicho bin

Luego la frecuencia se grafica utilizando barras. 

**Ejemplo:** Para construir el siguiente histograma se utilizaron $7$ bines de tamao $1$ regularmente distribuidos en el rango $[-3,4]$

<img src="https://phuijse.github.io/PythonBook/_images/stats4.png">

La principal utilidad del histograma es exploracin de las propiedades estadsticos de nuestros datos. En base al histograma podemos extraer informacin acerca de:

- La cantidad de modas
- La localizacin (media) y ancho (varianza) de las modas
- La diferencia relativa entre las colas de cada moda (simetra)

entre otras. Del ejemplo anterior podemos ver que la distribucin es unimodal, con ubicacin cercana (aunque ligeramente inferior) a cero y con una ligera asimetra de colas. 

:::{caution}

La cantidad de particiones (bines) utilizados para construir el histograma afecta de forma importante nuestras observaciones. Conviene probar con distintos valores de este hiperparmetro y comparar. Ms adelante ahondaremos en este punto.

:::

:::{note}

El ejemplo anterior corresponde al caso particular ms comn donde los cajones son disjuntos y de igual tamao. Otros mtodos como, [Bayesian blocks](https://www.astroml.org/examples/algorithms/plot_bayesian_blocks.html) puede utilizarse para construir histogramas con particiones de tamao variable 


:::



### Histograma con Python

Utilizando la librera Numpy podemos utilizar

```python
numpy.histogram(a, # arreglo (unidimensional) de datos
                bins=10, # Nmeros de particiones (cajones) o un arreglo con los lmites de las mismas
                range=None, # Tupla con valor inicial y final
                density=None, # Si es True entonces retorna densidades (frecuencia relativa)
                ...
               )
```

que retorna una tupla con dos arreglos: los bordes de las particiones y las frecuencias (o densidades) de las particiones. Veamos un ejemplo:

````python
from dists import GaussianMixture

# Generacin de los datos
dist = GaussianMixture(locs=[-4, 3], scales=[2, 2], weights=[0.7, 0.3])
samples = dist.rvs(1000, seed=1234)
x = np.linspace(np.amin(samples), np.amax(samples), num=1000)
pdf = dist.pdf(x)

# Construccin de histogramas con distintas particiones e igual rango
histograms = {}
for nbins in [1, 2, 5, 10, 20, 50, 100]:
    histograms[nbins] = np.histogram(samples, bins=nbins, 
                                     range=(np.amin(samples), np.amax(samples)), 
                                     density=True)
````

A continuacin se muestra el histograma de las muestras con la distribucin subyacente de fondo. 

> Utiliza el control deslizante para estudiar la influencia del nmero de particiones (bins):

````python
dist_plot = hv.Curve((x, pdf)).opts(line_width=2, color='k', width=500, height=350)
hmap = hv.HoloMap(kdims='Number of bins')
for key, (freqs, bins) in histograms.items():
    hmap[key] = hv.Histogram((bins, freqs), kdims='x', vdims='Density').opts(alpha=0.75)

hmap * dist_plot
````

:::{note}

En general

- Un nmero muy pequeo de bins "pierde" las caractersticas de la distribucin, por ejemplo la multimodalidad
- Un nmero muy grande de bins introduce ruido que dificulta la lectura del histograma

:::

### Cmo seleccionar el nmero de particiones?

Este hiperparamtro puede seleccionar haciendo una validacin cruzada sobre nuestros datos. Una mtrica ampliamente utilizada para lo anterior es el error cuadrtico medio asinttico (Asymptotic Mean Integrated Square Error, AMISE)

Tambin existen algunas "reglas del dedo" (*rules of thumb*), que suelen ser adecuadas como punto de partida. Dos ejemplos son la regla de Scott y la regla de Silverman. Ambas

- son proporcionales con la dispersin de los datos e inversamente proporcional con la cantidad de muestras
- se obtienen de imponer supuestos sobre la distribucin subyacente de los datos

Por ejemplo la regla de Silverman para el ancho de las particiones es:

$$
h = 0.9 \frac{\min[\sigma, 0.7412 (q_{75} - q_{25})]}{N^{1/5}},
$$

donde $N$ es el nmero de observaciones, $\sigma$ es la desviacin estndar y $q_{75}-q_{25}$ es el rango intercuartil. Luego usando este ancho podemos calcular el nmero de particiones con:

$$
N_{bins} = \frac{\max(x)-\min(x)}{h}
$$

:::{caution}

La regla de Silverman es ptima slo si la distribucin subyacente es Normal (Gaussiana). En caso general puede servir como punto de partida.

:::




## Kernel density estimation (KDE)

KDE puede considerarse como una alternativa al histograma donde cada muestra es su "propia particin", y donde las particiones pueden solaparse. Debido a lo anterior no se requiere escoger los lmites de las particiones, slo su ancho. 

La distribucin unidimensioanl aproximada con KDE para una muestra $\{x_i\}_{i=1,\ldots, N}$ es

$$
\hat f_h(x) = \frac{1}{Nh} \sum_{i=1}^N \kappa \left ( \frac{x - x_i}{h} \right)
$$

donde $h$ se llama **ancho de banda de kernel (kernel bandwidth)** y $\kappa(u)$ es una **funcin de kernel** que debe ser positiva, con media cero y adems integrar a la unidad.

Por ejemplo, un kernel muy utilizado es:

$$
\kappa(u) = \frac{1}{\sqrt{2\pi}} \exp \left ( - \frac{u^2}{2} \right),
$$

que se conoce como kernel Gaussiano. Otros kernels ampliamente utilizados en KDE son el kernel de Epanechnikov y el kernel Top-hat 

:::{caution}

Utilizar un kernel Gaussiano no implica a sumir que los datos se distribuyan Gaussianos (normales). Es totalmente correcto utilizar un kernel Gaussiano en datos no-Gaussianos

:::

### KDE en Python

KDE est implementado en el mdulo `sklearn.neighbors` como

```python
KernelDensity(kernel='gaussian', # kernel
              bandwidth=1.0, # ancho
              ...
             )
```

Sus mtodos ms importantes son

- `fit(xi: np.ndarray)`: Guarda las muestras que se utilizarn para calcular las densidades
- `score_samples(x: np.ndarray)`: Returna un arreglo de numpy con el logaritmo de las densidades evaluadas en el arreglo `x`

A continuacin se la estimacin con kernels de la distribucin del ejemplo anterior:

````python
from sklearn.neighbors import KernelDensity
hs = 0.9*np.std(samples)*len(samples)**(-1/5)

densities = {}
for k in [1/8, 1/4, 1/2, 1, 2, 4, 8]:
    for kernel in ["gaussian", "epanechnikov", "tophat"]:
        kde = KernelDensity(kernel=kernel, bandwidth=hs*k).fit(samples.reshape(-1, 1))
        densities[k, kernel] = np.exp(kde.score_samples(x.reshape(-1, 1)))
````

> Utiliza el control deslizante para estudiar la influencia de la funcin de kernel y de su ancho (indicado en k veces el ancho de Silverman)

````python
dist_plot = hv.Curve((x, pdf)).opts(line_width=2, color='k', width=500, height=350)
hmap = hv.HoloMap(kdims=['Width/Silverman', 'Kernel'])
for key, density in densities.items():
    hmap[key] = hv.Curve((x, density), kdims='x', vdims='Density').opts(line_width=2)

hmap * dist_plot
````

:::{note}

En general

- Un ancho grande las caractersticas de la distribucin
- Un ancho pequeo introduce ruido

:::
